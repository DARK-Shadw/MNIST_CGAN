```python
import torch
import torch.nn as nn
import torchvision.transforms as T
from torch.utils.data import DataLoader
import torchvision
```

```python
transform = T.Compose([
    T.ToTensor(),
    T.Normalize([0.5],[0.5])
])
```

```python
batch_size = 64
img_size = 28
num_classes = 10
latent_dim = 100
lr = 0.0002
epochs = 50
device = "cuda"
```

```python
data_loader = DataLoader(
    dataset= torchvision.datasets.MNIST(root='./data', train=True, transform=transform, download=True),
    shuffle = True,
    batch_size = batch_size,
)
```

```python
class Generator(nn.Module):
    def __init__(self, latent_dim, num_classes, img_size):
        super(Generator, self).__init__()
        self.label_emb = nn.Embedding(num_classes, num_classes)
        self.img_size = img_size
        self.model = nn.Sequential(
            nn.Linear(latent_dim + num_classes, 128),
            nn.ReLU(),
            nn.Linear(128, 256),
            nn.BatchNorm1d(256, 0.8),
            nn.ReLU(),
            nn.Linear(256, 512),
            nn.BatchNorm1d(512, 0.8),
            nn.ReLU(),
            nn.Linear(512, img_size * img_size),
            nn.Tanh(),
        ).to(device)
    def forward(self, noise, labels):
        gen_input = torch.cat((noise, self.label_emb(labels)), -1)
        img = self.model(gen_input)
        img = img.view(img.size(0), 1, self.img_size, self.img_size)
        return img
```

```python
class Discriminator(nn.Module):
    def __init__(self, num_classes, img_size):
        super(Discriminator, self).__init__()
        self.label_emb = nn.Embedding(num_classes, num_classes)
        self.model = nn.Sequential(
            nn.Linear(num_classes + img_size * img_size, 512),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(512, 256),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(256, 1),
            nn.Sigmoid()
        ).to(device)
    def forward(self, img, labels):
        img_flat = img.view(img.size(0), -1)
        d_in = torch.cat((img_flat, self.label_emb(labels)), -1)
        validity = self.model(d_in)
        return validity
    
```

```python
loss_fn = nn.BCELoss()

gen = Generator(latent_dim, num_classes, img_size).to(device)
dis = Discriminator(num_classes, img_size).to(device)

optim_g = torch.optim.Adam(gen.parameters(), lr=lr, betas=(0.5, 0.999))
optim_d = torch.optim.Adam(dis.parameters(), lr=lr, betas=(0.5, 0.999))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

# Create directory for sample images
os.makedirs("cgan_samples", exist_ok=True)

# Function to save generated images
def save_sample_images(epoch, latent_dim, num_classes):
    n_row = 10  # One row per class
    # Generate fixed noise
    z = torch.randn(num_classes * n_row, latent_dim).to(device)
    # Generate labels 0-9 for each row
    labels = torch.tensor([i for i in range(num_classes) for _ in range(n_row)]).to(device)
    
    gen.eval()
    with torch.no_grad():
        sample_imgs = gen(z, labels).cpu()
    gen.train()
    
    # Plot and save
    fig, axes = plt.subplots(nrows=num_classes, ncols=n_row, figsize=(12, 12))
    for i, ax in enumerate(axes.flatten()):
        img = sample_imgs[i].squeeze().numpy()
        ax.imshow(img, cmap='gray')
        ax.axis('off')
        ax.set_title(f"Label: {labels[i].item()}")
    
    plt.tight_layout()
    plt.savefig(f"cgan_samples/epoch_{epoch}.png")
    plt.close()
```

```python
for epoch in range(epochs):
    for i, (imgs, labels) in enumerate(data_loader):
        valid = torch.ones(imgs.size(0), 1).to(device)
        fake = torch.zeros(imgs.size(0), 1).to(device)

        optim_d.zero_grad()

        z = torch.randn(imgs.size(0), latent_dim).to(device)
        gen_labels = torch.randint(0, num_classes, (imgs.size(0), )).to(device)

        gen_imgs = gen(z, gen_labels).to(device)

        imgs = imgs.to(device)
        labels = labels.to(device)

        real_loss = loss_fn(dis(imgs, labels), valid)
        fake_loss = loss_fn(dis(gen_imgs.detach(), gen_labels), fake)

        dloss = (real_loss + fake_loss) / 2
        dloss.backward()

        optim_d.step()

        optim_g.zero_grad()

        gloss = loss_fn(dis(gen_imgs, gen_labels), valid)

        gloss.backward()
        optim_g.step()
    print(f"[Epoch {epoch}/{epochs}] [Batch {i}/{len(data_loader)}] [D loss: {dloss.item():.4f}] [G loss: {gloss.item():.4f}]")


    save_sample_images(epoch, latent_dim, num_classes)
```

```python
def generate_continuous_digit_sequence(gen, sequence, latent_dim=100, device="cuda", save_path=None):
    """
    Generate a continuous image of a sequence of digits with a black background using a trained CGAN generator.
    
    Args:
        gen: Trained Generator model
        sequence: A list of digits or an integer/string (which will be split into digits)
        latent_dim: Dimension of the latent space (default: 100)
        device: Device to run the model on (default: "cuda")
        save_path: If provided, saves the image to this path (default: None)
    
    Returns:
        A single continuous image containing all digits in the sequence with a black background
    """
    import torch
    import numpy as np
    import matplotlib.pyplot as plt
    
    # Convert to list of digits if input is an integer or string
    if isinstance(sequence, int) or isinstance(sequence, str):
        sequence = [int(digit) for digit in str(sequence)]
    
    # Generate one random noise vector per digit
    z = torch.randn(len(sequence), latent_dim).to(device)
    
    # Convert sequence to tensor
    labels = torch.tensor(sequence).to(device)
    
    # Generate images
    gen.eval()
    with torch.no_grad():
        images = gen(z, labels)
        images = images.cpu()
    gen.train()
    
    # Denormalize images from [-1, 1] to [0, 1]
    images = (images + 1) / 2
    
    # Create a continuous image by concatenating the digits horizontally
    img_list = [img.squeeze().numpy() for img in images]
    continuous_img = np.concatenate(img_list, axis=1)
    
    # Display the continuous image with a black background
    plt.figure(figsize=(len(sequence) * 1.5, 2))
    
    # Set figure background to black
    fig = plt.gcf()
    fig.patch.set_facecolor('black')
    
    # Create axes with black background
    ax = plt.gca()
    ax.set_facecolor('black')
    
    # Display the image
    plt.imshow(continuous_img, cmap='gray')
    plt.axis('off')
    
    plt.tight_layout(pad=0)
    
    if save_path:
        plt.savefig(save_path, bbox_inches='tight', pad_inches=0, facecolor='black')
        print(f"Image saved to {save_path}")
    
    plt.show()
    
    return continuous_img

# Example usage:
# After training your model:
# img = generate_continuous_digit_sequence(gen, 7396254321)
# or
# img = generate_continuous_digit_sequence(gen, [7, 3, 9, 6, 2, 5, 4, 3, 2, 1], save_path="sequence.png")
```
